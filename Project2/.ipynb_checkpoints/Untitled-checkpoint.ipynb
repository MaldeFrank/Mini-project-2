8. Split the aggregated data into five subsets by binning the attribute pH. Identify the subset with the highest density? What if you split the data in ten subsets?

```python
num_bins = 5


df_combined['pH_bins'] = pd.cut(df_combined['pH'], bins=num_bins)

grouped = df_combined.groupby('pH_bins')

interval_counts = grouped.size()


plt.figure(figsize=(10, 6))
plt.bar(interval_counts.index.astype(str), interval_counts.values, color='skyblue', edgecolor='black')
plt.xlabel('pH Bin Intervals')
plt.ylabel('Count')
plt.title('Histogram of pH Values with {} Bins'.format(num_bins))
plt.xticks(rotation=45)  
plt.grid(axis='y')  
plt.tight_layout()
plt.show()
```

```python
num_bins = 10


df_combined['pH_bins'] = pd.cut(df_combined['pH'], bins=num_bins)

grouped = df_combined.groupby('pH_bins')

interval_counts = grouped.size()


plt.figure(figsize=(10, 6))
plt.bar(interval_counts.index.astype(str), interval_counts.values, color='skyblue', edgecolor='black')
plt.xlabel('pH Bin Intervals')
plt.ylabel('Count')
plt.title('Histogram of pH Values with {} Bins'.format(num_bins))
plt.xticks(rotation=45) 
plt.grid(axis='y')  
plt.tight_layout()
plt.show()
```

13. Try to reduce the number of features of the aggregated data set by applying principal component analysis (PCA). What is the optimal number of components?


standardizing all values in dataset to see how many standard deviations away from the mean they are.

```python
scaler = StandardScaler()

df_standardized = pd.DataFrame(scaler.fit_transform(df_combined), columns=df_combined.columns)

df_standardized.head()
```

Computing covariance

```python
covariance_matrix = df_standardized.cov()

plt.figure(figsize=(10, 8))
sns.heatmap(covariance_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Covariance Matrix Heatmap')
plt.show()
```
